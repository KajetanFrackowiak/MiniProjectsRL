```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Flexibility of the model (x-axis)
flexibility = np.linspace(0, 10, 200)

# Simulated curves
bias_squared = (10 - flexibility)**2 / 10
variance = flexibility**2 / 15
bayes_error = np.full_like(flexibility, 1.5)  # irreducible error
training_error = bias_squared + 0.5 * variance  # less affected by variance
test_error = bias_squared + variance + bayes_error

# Plotting
plt.figure(figsize=(10,6))
plt.plot(flexibility, bias_squared, label='Bias', color='blue', linewidth=2)
plt.plot(flexibility, variance, label='Variance', color='red', linewidth=2)
plt.plot(flexibility, training_error, label='Training Error', color='green', linestyle='--', linewidth=2)
plt.plot(flexibility, test_error, label='Test Error', color='purple', linestyle='--', linewidth=2)
plt.plot(flexibility, bayes_error, label='Bayes Error', color='orange', linewidth=2)

plt.xlabel('Flexibility of the Method')
plt.ylabel('Error / Value')
plt.title('Bias-Variance Tradeoff')
plt.legend()
plt.grid(True)
plt.show()
```
